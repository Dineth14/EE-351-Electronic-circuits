# Interactive AC-DC Waveforms

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from ipywidgets import interactive
from power_electronics.converters.ac_dc.half_wave import HalfWaveRectifier
from power_electronics.converters.ac_dc.full_bridge import FullBridgeRectifier
from power_electronics.converters.ac_dc.scr_full_bridge import SCRFullBridgeRectifier
```

```python
def run(converter='full_bridge', Vac_peak=170.0, frequency=50.0, R_load=20.0, C_filter=0.002, alpha_deg=30.0):
    cls = {'half_wave': HalfWaveRectifier, 'full_bridge': FullBridgeRectifier, 'scr_full_bridge': SCRFullBridgeRectifier}[converter]
    params = {'Vac_peak': Vac_peak, 'frequency': frequency, 'R_load': R_load, 'L_filter': 1e-3, 'C_filter': C_filter, 'alpha_deg': alpha_deg}
    m = cls(params)
    r = m.simulate(0.2, 5000)
    keys=[k for k,v in r.signals.items() if getattr(v,'ndim',1)==1]
    fig,ax=plt.subplots(len(keys),1,figsize=(12,max(6,1.7*len(keys))),sharex=True)
    if len(keys)==1: ax=[ax]
    for a,k in zip(ax,keys): a.plot(r.time,r.signals[k]); a.set_ylabel(k); a.grid(True,alpha=0.2)
    ax[-1].set_xlabel('Time (s)'); plt.tight_layout()
    display(pd.DataFrame([r.metrics]).T.rename(columns={0:'value'}))
    th=m.get_theoretical_values(); display(pd.DataFrame({'Theoretical':th,'Simulated':{k:r.metrics.get(k,np.nan) for k in th}}))
```

```python
interactive(run, converter=['half_wave','full_bridge','scr_full_bridge'], Vac_peak=(100.0,325.0,1.0), frequency=(40.0,100.0,1.0), R_load=(5.0,100.0,1.0), C_filter=(0.0,0.01,0.0001), alpha_deg=(0.0,170.0,1.0));
```
