# Interactive DC-DC Waveforms

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from ipywidgets import interactive
from power_electronics.converters.dc_dc.buck import BuckConverter
from power_electronics.converters.dc_dc.boost import BoostConverter
from power_electronics.converters.dc_dc.buck_boost import BuckBoostConverter
from power_electronics.converters.dc_dc.flyback import FlybackConverter
```

```python
def run(converter='buck', Vin=48.0, duty_cycle=0.4, frequency=20000.0, R_load=15.0):
    cls={'buck':BuckConverter,'boost':BoostConverter,'buck_boost':BuckBoostConverter,'flyback':FlybackConverter}[converter]
    p={'Vin':Vin,'duty_cycle':duty_cycle,'frequency':frequency,'L':1e-3,'C':220e-6,'R_load':R_load,'R_L':0.05,'n':2.0}
    m=cls(p); r=m.simulate(0.05,7000)
    keys=[k for k,v in r.signals.items() if getattr(v,'ndim',1)==1]
    fig,ax=plt.subplots(len(keys),1,figsize=(12,max(6,1.6*len(keys))),sharex=True)
    if len(keys)==1: ax=[ax]
    for a,k in zip(ax,keys): a.plot(r.time,r.signals[k]); a.set_ylabel(k); a.grid(True,alpha=0.2)
    ax[-1].set_xlabel('Time (s)'); plt.tight_layout()
    display(pd.DataFrame([r.metrics]).T.rename(columns={0:'value'}))
    th=m.get_theoretical_values(); display(pd.DataFrame({'Theoretical':th,'Simulated':{k:r.metrics.get(k,np.nan) for k in th}}))
```

```python
interactive(run, converter=['buck','boost','buck_boost','flyback'], Vin=(12.0,100.0,1.0), duty_cycle=(0.05,0.95,0.01), frequency=(5000.0,100000.0,1000.0), R_load=(2.0,100.0,1.0));
```
