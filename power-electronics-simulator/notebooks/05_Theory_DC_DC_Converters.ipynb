# DC-DC Converter Theory

```python
import numpy as np
import matplotlib.pyplot as plt
import schemdraw
import schemdraw.elements as elm
```

```python
with schemdraw.Drawing() as d:
    d += elm.Battery().label('Vin')
    d += elm.Switch().right().label('Q')
    d += elm.Inductor().right().label('L')
    d += elm.Capacitor().down().label('C')
    d += elm.Resistor().down().label('R')
d
```

- Buck: $V_o=DV_{in}$
- Boost: $V_o=V_{in}/(1-D)$
- Buck-boost: $V_o=-DV_{in}/(1-D)$
- Volt-second balance: $\int_0^{T_s}v_Ldt=0$
- Charge balance: $\int_0^{T_s}i_Cdt=0$

| Converter | Gain | Polarity |
|---|---|---|
| Buck | D | + |
| Boost | 1/(1-D) | + |
| Buck-boost | -D/(1-D) | - |
| Cuk | -D/(1-D) | - |
| SEPIC | D/(1-D) | + |

```python
d=np.linspace(0.05,0.95,80); eta=94-5*(d-0.5)**2
load=np.linspace(0.2,1.0,80); eta2=86+13*load-4*load**2
plt.figure(figsize=(8,4)); plt.plot(d,eta,label='vs duty'); plt.plot(load,eta2,label='vs load'); plt.grid(True,alpha=0.3); plt.legend(); plt.ylabel('Efficiency (%)')
```
