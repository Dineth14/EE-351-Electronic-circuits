# Interactive DC-AC Waveforms

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from ipywidgets import interactive
from power_electronics.converters.dc_ac.single_phase_hbridge import SinglePhaseHBridgeVSI
from power_electronics.converters.dc_ac.three_phase_vsi import ThreePhaseVSI
from power_electronics.converters.dc_ac.npc_inverter import NPCInverter
```

```python
def run(converter='single_phase_hbridge', Vdc=300.0, f_output=50.0, f_carrier=5000.0, modulation_index=0.8):
    cls={'single_phase_hbridge':SinglePhaseHBridgeVSI,'three_phase_vsi':ThreePhaseVSI,'npc_inverter':NPCInverter}[converter]
    p={'Vdc':Vdc,'f_output':f_output,'f_carrier':f_carrier,'modulation_index':modulation_index,'R_load':20.0,'L_load':0.01,'C_load':0.0,'control_mode':'SPWM'}
    m=cls(p); r=m.simulate(0.1,5000)
    keys=[k for k,v in r.signals.items() if getattr(v,'ndim',1)==1]
    fig,ax=plt.subplots(len(keys),1,figsize=(12,max(6,1.6*len(keys))),sharex=True)
    if len(keys)==1: ax=[ax]
    for a,k in zip(ax,keys): a.plot(r.time,r.signals[k]); a.set_ylabel(k); a.grid(True,alpha=0.2)
    ax[-1].set_xlabel('Time (s)'); plt.tight_layout()
    display(pd.DataFrame([r.metrics]).T.rename(columns={0:'value'}))
    th=m.get_theoretical_values(); display(pd.DataFrame({'Theoretical':th,'Simulated':{k:r.metrics.get(k,np.nan) for k in th}}))
```

```python
interactive(run, converter=['single_phase_hbridge','three_phase_vsi','npc_inverter'], Vdc=(50.0,800.0,1.0), f_output=(10.0,100.0,1.0), f_carrier=(1000.0,20000.0,100.0), modulation_index=(0.1,1.0,0.01));
```
