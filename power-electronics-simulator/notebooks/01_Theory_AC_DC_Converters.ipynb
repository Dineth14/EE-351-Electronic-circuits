## Equations
- $V_{dc,half}=\frac{V_m}{\pi}$
- $V_{dc,full}=\frac{2V_m}{\pi}$
- $V_{dc,SCR}=\frac{2V_m}{\pi}\cos\alpha$
- Ripple: $\Delta V_o \approx \frac{I_o}{f_r C}$

| Converter | Devices | Control | PF |
|---|---:|---|---:|
| Half-wave | 1 | No | Low |
| Full-bridge | 4 | No | Medium |
| SCR bridge | 4 SCR | Yes | Medium |
| Vienna | 6 + active | Yes | High |
