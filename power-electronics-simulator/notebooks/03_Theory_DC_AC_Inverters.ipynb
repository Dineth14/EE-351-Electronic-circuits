```python
d=np.linspace(0.1,0.95,80); eta=95-3*(d-0.7)**2*10
load=np.linspace(0.2,1.0,80); eta2=90+8*load-2.5*load**2
plt.figure(figsize=(8,4)); plt.plot(d,eta,label='vs duty'); plt.plot(load,eta2,label='vs load'); plt.grid(True,alpha=0.3); plt.legend(); plt.ylabel('Efficiency (%)')
```
